# **Gender distribution according to the target variable**

```python
 plt.title('Gender distribution according to the target variable')
 sns.countplot(data= df, x= 'sex', hue= 'target')
 plt.xticks([0,1], ['Female','Male'])
 plt.legend(labels= ['No-Disease', 'Disease'])
 plt.show()
```

# **Age Distribution in The Dataset**

# **Chest Pain Type as Per Gender Variable**

```python
df.columns
```

```python
plt.title('Chest Pain Type as Per Gender')
sns.countplot(data= df, x= 'cp', hue= 'sex')
plt.xticks([0,1,2,3], ['Typical Angina', 'Atypical Angina', 'Non-anginal Pain', 'Asymptomatic'])
plt.legend(labels= ['Female', 'Male'])
plt.show()
```

# **Chest Pain Type as Per Target Variable**

```python
plt.title('Chest Pain Type as Per Target')
sns.countplot(data= df, x= 'cp', hue= 'target')
plt.xticks([0,1,2,3], ['Typical Angina', 'Atypical Angina', 'Non-anginal Pain', 'Asymptomatic'])
plt.legend(labels= ['Non-Disease', 'Disease'])
plt.show()
```

# **Compare Resting Blood Presure as Per Sex Column**
